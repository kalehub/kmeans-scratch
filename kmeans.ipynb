```python
import numpy as np
dataset_dir = 'dataset-cluster.txt'
k = 3
```

```python
# read dataset
dataset = np.loadtxt(dataset_dir)
dataset
```

```python
# memilih centroid acak
centroid = dataset[np.random.choice(dataset.shape[0], k, replace=False)]
centroid
```

```python
# subtraction in numpy
print('shape dataset : ',dataset.shape)
centroid_arr = np.matrix(centroid)
euc_matrix = np.ndarray(shape=(dataset.shape[0],0))
print('shape euc_matrix :',euc_matrix.shape)
for i in range(0, k):
    centro = np.repeat(centroid_arr[i,:], dataset.shape[0],axis=0)
    subtr = abs(np.subtract(dataset,centro))
    euc = np.sqrt(np.square(subtr).sum(axis=1))
    euc_matrix = np.concatenate((euc_matrix, euc),axis=1)

cluster_matrix = np.ravel(np.argmin(euc_matrix,axis=1))
list_cluster = [[] for x in range(k)]
for i in range(0, dataset.shape[0]):
    list_cluster[np.asscalar(cluster_matrix[i])].append(dataset[i,:])

print(list_cluster[:1][0])
    
    
    
    
```
